Task 4(Online delivery)

```python
# Determine the percentage of restaurants that offer online delivery

Total_restaurant = df['Restaurant Name'].count()

Categorize_online_delivery = df.groupby('Has Online delivery')['Restaurant Name'].count().reset_index()
Categorize_online_delivery.columns = ['Has Online delivery','Restaurant Count']

Online_delivery_yes = Categorize_online_delivery[Categorize_online_delivery['Has Online delivery'] == 'Yes']

Restaurant_count_yes = Online_delivery_yes['Restaurant Count'].iloc[0]

Percentage_online_delivery = (Restaurant_count_yes / Total_restaurant )*100

print(f"Percentage of restaurants that offer online delivery: {Percentage_online_delivery:.2f}%")
```

Task 3(Price Range vs. Online Delivery and Table Booking)

```python
#Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

online_booking_crosstab = pd.crosstab(df['Price range'], df['Has Online delivery'], normalize='index') * 100
table_availability_crosstab = pd.crosstab(df['Price range'], df['Has Table booking'], normalize='index') * 100

# Plotting the relationship between price range and online bookings
plt.figure(figsize=(8,4))
sns.heatmap(online_booking_crosstab, annot=True, cmap="Blues", cbar_kws={'label': 'Percentage (%)'})
plt.title('Relationship Between Price Range and Online Bookings')
plt.ylabel('Price Range')
plt.xlabel('Online Bookings')
plt.show()

# Plotting the relationship between price range and table availability
plt.figure(figsize=(8,4))
sns.heatmap(table_availability_crosstab, annot=True, cmap="Greens", cbar_kws={'label': 'Percentage (%)'})
plt.title('Relationship Between Price Range and Table Availability')
plt.ylabel('Price Range')
plt.xlabel('Table Availability')
plt.show()
```

```python
#Determine if higher-priced restaurants are more likely to offer these services.

from scipy.stats import chi2_contingency

online_booking_contingency = pd.crosstab(df['Price range'], df['Has Online delivery'])
table_availability_contingency = pd.crosstab(df['Price range'], df['Has Table booking'])

# Chi-square test for online booking
chi2_online, p_online, dof_online, ex_online = chi2_contingency(online_booking_contingency)
print(f'Chi-square test for Online Bookings: chi2={chi2_online:.4f}, p-value={p_online:.4f}')

# Chi-square test for table availability
chi2_table, p_table, dof_table, ex_table = chi2_contingency(table_availability_contingency)
print(f'Chi-square test for Table Availability: chi2={chi2_table:.4f}, p-value={p_table:.4f}')
```
